```python
import pandas as pd
import numpy as np
import statistics as s
df=pd.read_csv(r"C:\Users\Arti\Desktop\mtcars.csv")
df
```

```python

def color_change(val):
    
    
    if (val in df.qsec.values):
      m1=np.mean(df.qsec)
      #m1
      std1=s.stdev(df.qsec)
      #std1
      U_LV=m1-std1#given data is very normal,all three color would not be displayed,therefore standard deviation is not multiplied with 2
      
      U_UV=m1+std1
      
      if (val < U_LV):
       color = 'red' 
       return ('color: %s' % color)
      
      if (val> U_UV):
       color ='green'
       return ('color: %s' % color)
      
      if( val>U_LV) and (val<U_UV):
       color ='blue'
       return ('color: %s' % color)

    if (val in df.disp.values):
      m1=np.mean(df.disp)
      
      std1=s.stdev(df.disp)
      
      U_LV=m1-std1
      
      U_UV=m1 + std1
      
      if (val < U_LV):
       color = 'red' 
       return ('color: %s' % color)
      
      if (val> U_UV):
       color ='green'
       return ('color: %s' % color)
    
      if( val>U_LV) and (val<U_UV):
       color ='blue'
       return ('color: %s' % color)

    if (val in df.mpg.values):
      m1=np.mean(df.mpg)
      #m1
      std1=s.stdev(df.mpg)
      #std1
      U_LV=m1 - std1
      
      U_UV=m1 + std1
      
      if (val < U_LV):
       color = 'red' 
       return ('color: %s' % color)
      
      if (val> U_UV):
       color ='green'
       return ('color: %s' % color)
      
      if( val>U_LV) and (val<U_UV):
       color ='blue'
       return ('color: %s' % color)
    
    

    

    if (val in df.drat.values):
      m1=np.mean(df.drat)
      #m1
      std1=s.stdev(df.drat)
      #std1
      O_LV=m1-std1 
      
      O_UV=m1+std1
      
      if (val < O_LV):
       color = 'red' 
       return ('color: %s' % color)
      
      if (val> O_UV):
       color ='green'
       return ('color: %s' % color)
      
      if( val>O_LV) and (val<O_UV):
       color ='blue'
       return ('color: %s' % color)

    if (val in df.wt.values):
      m1=np.mean(df.wt)
      #m1
      std1=s.stdev(df.wt)
      #std1
      O_LV=m1 - std1
      
      O_UV=m1 + std1
      
      if (val < O_LV):
       color = 'red' 
       return ('color: %s' % color)
      
      if (val > O_UV):
       color ='green'
       return ('color: %s' % color)
      
      if( val>O_LV) and (val<O_UV):
       color ='blue'
       return ('color: %s' % color)
    
    if (val in df.hp.values):
      m1=np.mean(df.hp)#mean of the column
      #m1
      std1=s.stdev(df.hp)#standard deviation of the column
      #std1
      O_LV=m1-std1
      
      O_UV=m1+std1
      
      if (val < O_LV):
       color = 'red' 
       return ('color: %s' % color)
       
      if (val > O_UV):
       color ='green'
       return ('color: %s' % color)
      
      if( val>O_LV) & (val<O_UV):
       color ='blue'
       return ('color: %s' % color)
    
    
    
dz=df.style.applymap(color_change,subset=pd.IndexSlice[['disp','qsec','mpg','hp','drat','wt']])
dz
```
